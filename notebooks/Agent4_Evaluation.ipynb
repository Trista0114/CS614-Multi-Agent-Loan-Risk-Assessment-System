```python
import json
import glob
import os

folder_path = '/content/drive/MyDrive/Gen AI group project/reports'

# Based on your example "report_applicant_3.json", we use "report_applicant_*.json"
# This will grab all files starting with "report_applicant_" and ending with ".json"
file_pattern = os.path.join(folder_path, 'report_applicant_*.json')

# Output filename for the merged file
output_filename = os.path.join(folder_path, 'merge.json')

# Prepare an empty list to store all data
all_data = []

# --- 3. Start searching and merging ---
print(f"Searching path: {file_pattern}")

file_list = glob.glob(file_pattern)

if not file_list:
    print(f"Error: No 'report_applicant_*.json' files found in '{folder_path}'.")
    print("Please check your path and filenames.")
else:
    print(f"Found {len(file_list)} files, starting merge...")

    for file_name in file_list:
        # Avoid reading our own output file if it already exists
        if file_name == output_filename:
            continue

        print(f"  > Reading: {os.path.basename(file_name)}")
        try:
            with open(file_name, 'r', encoding='utf-8') as f:
                data = json.load(f)

                # Scenario 1: Add the entire file's content (one object) to the list
                all_data.append(data)

        except json.JSONDecodeError:
            print(f"    Error: File {os.path.basename(file_name)} is not valid JSON.")
        except Exception as e:
            print(f"    Error processing file {os.path.basename(file_name)}: {e}")

    # --- 4. Write the merged file ---
    if all_data: # Only write if we have data
        print(f"\nWriting {len(all_data)} records to {output_filename} ...")

        with open(output_filename, 'w', encoding='utf-8') as f:
            # ensure_ascii=False ensures non-ASCII characters (like Chinese) display correctly
            # indent=4 formats the JSON to be human-readable
            json.dump(all_data, f, ensure_ascii=False, indent=4)

        print("Merge complete!")
    else:
        print("No data was merged.")
```

```python
import pandas as pd
import re
from IPython.display import display # For pretty printing DataFrames in Notebooks

def evaluate_agent4_comprehensively(df: pd.DataFrame):
    """
    Performs a comprehensive evaluation of Agent 4's output.

    This function merges:
    1. Decision Consistency:
       - 'final_decision' variable vs 'reasoning_trace'
       - 'reasoning_trace' vs the first word of 'recommendation'
       - 'final_decision' variable vs the first word of 'recommendation'
    2. Recommendation Faithfulness:
       - Hallucination: Checks if 'PASS' agents are incorrectly reported as 'failed'.
       - Missing Fail: Checks if 'FAIL' agents are completely omitted from the 'recommendation'.
    3. FAIL Case Completeness - [Your requested feature]:
       - For all 'Disapproved' cases, checks if the *first sentence* of the
         'recommendation' accurately and completely lists all agents that *truly* failed.
    """

    print("=" * 70)
    print("Agent 4 Comprehensive Evaluation Report")
    print("=" * 70)

    # --- 1. Initialize Counters ---
    total_cases = len(df)
    disapproved_cases_count = 0 # Counter for total disapproved cases

    # Decision Consistency Errors
    errors_decision_trace = 0
    errors_trace_recommendation = 0
    errors_decision_recommendation = 0

    # Recommendation Faithfulness Errors
    errors_hallucination = 0
    errors_fail_missing = 0

    # Fail Case Completeness Errors (Your requested feature)
    errors_fail_case_completeness = 0

    case_results = [] # Stores detailed errors for each case

    if total_cases == 0:
        print("No cases found in the DataFrame.")
        return

    # --- 2. Iterate through all cases ---
    for index, row in df.iterrows():
        applicant_id = row.get('applicant_data.ID', index)
        report = row.get('agent4_final_report.recommendation', "")
        decision = row.get('agent4_final_report.final_decision', "N/A")
        trace = row.get('agent4_final_report.reasoning_trace', [])

        case_errors_list = []

        # --- Check 1: Decision Consistency (All cases) ---
        if trace:
            trace_decision = trace[-1].split(': ')[-1]
            if decision != trace_decision:
                errors_decision_trace += 1
                case_errors_list.append(f"Decision-Trace Mismatch (State: '{decision}', Trace: '{trace_decision}')")

            recommendation_first_word = ""
            if report:
                try:
                    first_word_raw = report.split()[0]
                    recommendation_first_word = first_word_raw.strip(".,!?:")
                except IndexError:
                    recommendation_first_word = "[EMPTY REPORT]"

            if recommendation_first_word != trace_decision:
                errors_trace_recommendation += 1
                case_errors_list.append(f"Trace-Recommendation Mismatch (Trace: '{trace_decision}', Rec: '{recommendation_first_word}')")

        if not report.startswith(decision):
            errors_decision_recommendation += 1
            case_errors_list.append(f"Recommendation-Decision Mismatch (State: '{decision}', Rec starts with: '{recommendation_first_word}')")

        # --- Check 2: Recommendation Faithfulness (All cases) ---
        a3_status = row.get('agent3_result.Overall_compliance_status', 'PASS')
        a1_trace = next((t for t in trace if "Credit Risk" in t), None)
        a2_trace = next((t for t in trace if "Fraud Risk" in t), None)

        # Check 2a: Hallucination
        hallucination_found = False
        if a3_status == 'PASS' and 'Agent 3 failed' in report:
             hallucination_found = True
             case_errors_list.append("Hallucination: A3 PASSED, but report implies it FAILED.")
        if a1_trace and 'PASS' in a1_trace and ('Agent 1 failed' in report or 'Agent 1 (Credit Risk) failed' in report):
             hallucination_found = True
             case_errors_list.append("Hallucination: A1 PASSED, but report implies it FAILED.")
        if a2_trace and 'PASS' in a2_trace and ('Agent 2 failed' in report or 'Agent 2 (Fraud Risk) failed' in report):
             hallucination_found = True
             case_errors_list.append("Hallucination: A2 PASSED, but report implies it FAILED.")
        if hallucination_found:
            errors_hallucination += 1

        # Check 2b: Missing Fail - Checks *entire* report
        fail_missing_found = False
        if a3_status == 'FAIL' and not re.search(r'Agent 3|Compliance', report, re.IGNORECASE):
             fail_missing_found = True
             case_errors_list.append("Faithfulness Error: A3 FAILED, but NOT mentioned in *any* part of the recommendation.")
        if a1_trace and 'FAIL' in a1_trace and not re.search(r'Agent 1|Credit', report, re.IGNORECASE):
             fail_missing_found = True
             case_errors_list.append("Faithfulness Error: A1 FAILED, but NOT mentioned in *any* part of the recommendation.")
        if a2_trace and 'FAIL' in a2_trace and not re.search(r'Agent 2|Fraud', report, re.IGNORECASE):
             fail_missing_found = True
             case_errors_list.append("Faithfulness Error: A2 FAILED, but NOT mentioned in *any* part of the recommendation.")
        if fail_missing_found:
             errors_fail_missing += 1

        # --- Check 3: FAIL Case Completeness (Only for 'Disapproved' cases) ---
        if decision == 'Disapproved':
            disapproved_cases_count += 1

            # Find failures from trace
            failed_agents_from_trace = set()
            for trace_line in trace:
                if "FAIL" in trace_line:
                    if "Compliance" in trace_line:
                        failed_agents_from_trace.add("Agent 3")
                    elif "Credit Risk" in trace_line:
                        failed_agents_from_trace.add("Agent 1")
                    elif "Fraud Risk" in trace_line:
                        failed_agents_from_trace.add("Agent 2")

            # Find mentions in *first sentence*
            first_sentence = report.split('.')[0]
            agents_mentioned_in_first_sentence = set()
            if re.search(r'Agent 1|Credit', first_sentence, re.IGNORECASE):
                agents_mentioned_in_first_sentence.add("Agent 1")
            if re.search(r'Agent 2|Fraud', first_sentence, re.IGNORECASE):
                agents_mentioned_in_first_sentence.add("Agent 2")
            if re.search(r'Agent 3|Compliance', first_sentence, re.IGNORECASE):
                agents_mentioned_in_first_sentence.add("Agent 3")

            # Compare
            missing_agents = failed_agents_from_trace - agents_mentioned_in_first_sentence

            if missing_agents:
                errors_fail_case_completeness += 1
                case_errors_list.append(f"FailCaseCompleteness Error: First sentence MISSED: {missing_agents}")

        case_results.append({
            "id": applicant_id,
            "errors": ", ".join(case_errors_list) if case_errors_list else "None"
        })

    # --- 3. Generate Final Summary Report ---

    print(f"Analyzed {total_cases} total cases.\n")

    # Standard 1: Decision Consistency
    total_consistency_errors = errors_decision_trace + errors_trace_recommendation + errors_decision_recommendation
    consistency_accuracy = (total_cases - total_consistency_errors) / total_cases if total_cases > 0 else 1
    print(f"--- 1. Decision Consistency ---")
    print(f"  Result: {consistency_accuracy:.1%} Correct ({total_cases - total_consistency_errors}/{total_cases} cases)")
    if total_consistency_errors > 0:
        print(f"    - Decision vs Trace Errors: {errors_decision_trace} cases")
        print(f"    - Trace vs Recommendation Errors: {errors_trace_recommendation} cases")
        print(f"    - Decision vs Recommendation Errors: {errors_decision_recommendation} cases")

    # Standard 2: Recommendation Faithfulness (General)
    total_faithfulness_errors = errors_hallucination + errors_fail_missing
    faithfulness_accuracy = (total_cases - total_faithfulness_errors) / total_cases if total_cases > 0 else 1
    print(f"\n--- 2. Recommendation Faithfulness ---")
    print(f"  Result: {faithfulness_accuracy:.1%} Correct ({total_cases - total_faithfulness_errors}/{total_cases} cases)")
    if total_faithfulness_errors > 0:
        print(f"    - Hallucination Errors: {errors_hallucination} cases")
        print(f"    - Missing Fail Errors: {errors_fail_missing} cases")

    # Standard 3: FAIL Case Completeness (Your new check)
    print(f"\n--- 3. FAIL Case Completeness ---")
    print(f"  (This check only applies to the {disapproved_cases_count} 'Disapproved' cases)")
    if disapproved_cases_count > 0:
        completeness_accuracy = (disapproved_cases_count - errors_fail_case_completeness) / disapproved_cases_count
        print(f"  Result: {completeness_accuracy:.1%} Correct ({disapproved_cases_count - errors_fail_case_completeness}/{disapproved_cases_count} cases)")
        if errors_fail_case_completeness > 0:
            print(f"    - First Sentence Omission Errors: {errors_fail_case_completeness} cases")
    else:
        print("  Result: N/A (No 'Disapproved' cases)")

    # --- 4. Display Error Details ---
    print("\n" + "=" * 70)
    print("Error Case Details")
    print("=" * 70)

    df_errors = pd.DataFrame(case_results)
    df_errors = df_errors[df_errors['errors'] != "None"]

    if df_errors.empty:
        print("\n✓ Congratulations! All cases passed all checks.")
    else:
        # 'display' is nicer in notebooks
        display(df_errors.set_index('id'))
```

```python
import numpy as np # Make sure numpy is imported

def evaluate_recommendation_statistics(df: pd.DataFrame):
    """
    Analyzes the length and keyword frequency of the 'recommendation' field.
    """
    print("=" * 70)
    print("Agent 4 Recommendation Statistics")
    print("=" * 70)

    # Ensure we only select the 'recommendation' field
    recommendations = df['agent4_final_report.recommendation'].fillna("").tolist()

    if not recommendations:
        print("No recommendations found to analyze.")
        return

    total_cases = len(recommendations)

    # --- 1. Length Statistics ---
    # Calculate word count for each recommendation
    word_counts = [len(text.split()) for text in recommendations if text]

    if not word_counts:
        print("No non-empty recommendations found.")
        return

    print("Recommendation Length Statistics (words):")
    print(f"  Mean: {np.mean(word_counts):.1f} words")
    print(f"  Min:  {np.min(word_counts)} words")
    print(f"  Max:  {np.max(word_counts)} words")
    print("")

    # --- 2. Keyword Mention Frequency ---
    a1_mentions = 0
    a2_mentions = 0
    a3_mentions = 0

    for text in recommendations:
        lower_text = text.lower() # Convert to lowercase for case-insensitive matching

        # Check for Agent 1 mentions ( "Agent 1" or "credit" )
        if "agent 1" in lower_text or "credit" in lower_text:
            a1_mentions += 1

        # Check for Agent 2 mentions ( "Agent 2" or "fraud" )
        if "agent 2" in lower_text or "fraud" in lower_text:
            a2_mentions += 1

        # Check for Agent 3 mentions ( "Agent 3" or "compliance" )
        if "agent 3" in lower_text or "compliance" in lower_text:
            a3_mentions += 1

    print("Agent Mention Frequency in Recommendation:")
    print(f"  Agent 1 (Credit Risk): {a1_mentions}/{total_cases} ({a1_mentions/total_cases:.1%})")
    print(f"  Agent 2 (Fraud Risk):  {a2_mentions}/{total_cases} ({a2_mentions/total_cases:.1%})")
    print(f"  Agent 3 (Compliance):  {a3_mentions}/{total_cases} ({a3_mentions/total_cases:.1%})")
    print("")
```

```python
import json
import pandas as pd
import numpy as np
import re #
from typing import List, Dict, Any

# [ Cell 3: 已修正的 load_data ]
def load_data(filename: str = "merge.json") -> pd.DataFrame:
    """載入並扁平化 merge.json 檔案"""
    print(f"Loading data from {filename}...")
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            data = json.load(f)


        df = pd.json_normalize(
            data,
            meta=[
                ['applicant_data', 'ID'],
                ['applicant_data', 'Age'],
                ['agent1_result', 'analysis', 'reasoning'],
                ['agent1_result', 'analysis', 'risk_level'],
                ['agent1_result', 'analysis', 'default_probability'],
                ['agent1_result', 'reasoning'],
                ['agent2_result', 'reasoning'],
                ['agent2_result', 'risk_level'],
                ['agent2_result', 'fraud_probability'],
                ['agent3_result', 'Overall_compliance_status'],
                ['agent3_result', 'check_details'],
                ['agent4_final_report', 'recommendation']
            ]
        )


        df = df.rename(columns={
            'applicant_data.ID': 'applicant_data.ID',
            'agent1_result.analysis.reasoning': 'agent1_result.analysis.reasoning',
            'agent1_result.analysis.risk_level': 'agent1_result.analysis.risk_level',
            'agent1_result.analysis.default_probability': 'agent1_result.analysis.default_probability',
            'agent1_result.reasoning': 'agent1_result.reasoning',
            'agent2_result.reasoning': 'agent2_result.reasoning',
            'agent2_result.risk_level': 'agent2_result.risk_level',
            'agent2_result.fraud_probability': 'agent2_result.fraud_probability',
            'agent3_result.Overall_compliance_status': 'agent3_result.Overall_compliance_status',
            'agent3_result.check_details': 'agent3_result.check_details',
            'agent4_final_report.recommendation': 'agent4_final_report.recommendation'
        })


        print(f"✓ Loaded and normalized {len(df)} test cases.\n")
        return df
    except FileNotFoundError:
        print(f"✗ ERROR: File not found: {filename}")
        print("!!! make sure merge.json path correct !!!")
        return pd.DataFrame()
    except Exception as e:
        print(f"✗ ERROR loading file: {e}")
        traceback.print_exc()
        return pd.DataFrame()



def call_llm_judge(prompt_text: str) -> str:
    """使用 Qwen2.5 LLM Judge 生成回應"""
    if 'llm_judge_model' not in globals():
        print("✗ LLM Judge model not loaded.")
        return ""


    messages = [
        {"role": "user", "content": prompt_text}
    ]


    text = llm_judge_tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    model_inputs = llm_judge_tokenizer([text], return_tensors="pt").to(llm_judge_model.device)

    generated_ids = llm_judge_model.generate(
        model_inputs.input_ids,
        max_new_tokens=300,
        temperature=0.01,
        do_sample=False
    )


    response_ids = generated_ids[0][model_inputs.input_ids.shape[1]:]
    response = llm_judge_tokenizer.decode(response_ids, skip_special_tokens=True)

    return response

def extract_json_from_response(text: str) -> Dict[str, Any]:

    try:
        match = re.search(r'\{.*\}', text, re.DOTALL)
        if match:
            json_str = match.group(0)
            return json.loads(json_str)
        else:
            return {"error": "No JSON object found", "raw_response": text}
    except json.JSONDecodeError:
        return {"error": "Failed to decode JSON", "raw_response": text}
    except Exception as e:
         return {"error": str(e), "raw_response": text}
```

```python
# [ Cell 3: correct load_data ]
def load_data(filename: str = "merge.json") -> pd.DataFrame:
    print(f"Loading data from {filename}...")
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            data = json.load(f)

        df = pd.json_normalize(
            data,
            meta=[
                ['applicant_data', 'ID'],
                ['applicant_data', 'Age'],
                ['agent1_result', 'analysis', 'reasoning'],
                ['agent1_result', 'reasoning'],
                ['agent2_result', 'reasoning'],
                ['agent3_result', 'Overall_compliance_status'],
                ['agent3_result', 'check_details'],
                ['agent4_final_report', 'recommendation']
            ]
        )

        df = df.rename(columns={
            'applicant_data.ID': 'applicant_data.ID',
            'agent1_result.analysis.reasoning': 'agent1_result.analysis.reasoning',
            'agent1_result.reasoning': 'agent1_result.reasoning',
            'agent2_result.reasoning': 'agent2_result.reasoning',
            'agent3_result.Overall_compliance_status': 'agent3_result.Overall_compliance_status',
            'agent3_result.check_details': 'agent3_result.check_details',
            'agent4_final_report.recommendation': 'agent4_final_report.recommendation'
        })

        print(f"✓ Loaded and normalized {len(df)} test cases.\n")
        return df
    except FileNotFoundError:
        print(f"✗ ERROR: File not found: {filename}")
        print("!!! make sure merge.json path correct !!!")
        return pd.DataFrame()
    except Exception as e:
        print(f"✗ ERROR loading file: {e}")
        traceback.print_exc()
        return pd.DataFrame()



def call_llm_judge(prompt_text: str) -> str:
    """使用 Qwen2.5 LLM Judge 生成回應"""
    if 'llm_judge_model' not in globals():
        print("✗ LLM Judge model not loaded.")
        return ""


    messages = [
        {"role": "user", "content": prompt_text}
    ]


    text = llm_judge_tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    model_inputs = llm_judge_tokenizer([text], return_tensors="pt").to(llm_judge_model.device)

    generated_ids = llm_judge_model.generate(
        model_inputs.input_ids,
        max_new_tokens=300,
        temperature=0.01,
        do_sample=False
    )

    response_ids = generated_ids[0][model_inputs.input_ids.shape[1]:]
    response = llm_judge_tokenizer.decode(response_ids, skip_special_tokens=True)

    return response

def extract_json_from_response(text: str) -> Dict[str, Any]:

    try:

        match = re.search(r'\{.*\}', text, re.DOTALL)
        if match:
            json_str = match.group(0)
            return json.loads(json_str)
        else:
            return {"error": "No JSON object found", "raw_response": text}
    except json.JSONDecodeError:
        return {"error": "Failed to decode JSON", "raw_response": text}
    except Exception as e:
         return {"error": str(e), "raw_response": text}
```

```python
# [ Cell 4: Revised LLM Judge Function (with A1/A2 Status) ]

# --- [ NEW ENGLISH PROMPT TEMPLATE (v5) ] ---
# (Added A1/A2 Status fields and Status interpretation rules)
LLM_JUDGE_PROMPT_TEMPLATE = """
You are an expert LLM evaluator (LLM Judge). Your task is to strictly evaluate whether the "Agent 4 Recommendation" is fully and faithfully based on the "Original Reasoning" from Agents 1, 2, and 3.

Evaluate the "Recommendation" against the "Original Reasoning" (Context) provided below.

**Criteria:**
1.  **Faithful:** Every claim about an agent's findings in the "Recommendation" MUST be clearly supported by that agent's "Original Reasoning" (including the provided Status, Probability, and Reasoning Text).
2.  **No Hallucination:** The "Recommendation" MUST NOT invent information or reasons not present in the "Original Reasoning".
3.  **No Contradiction:** The "Recommendation" MUST NOT contradict the "Original Reasoning".
4.  **FAIL Case Check:** If the final decision is "Disapproved", the "Recommendation" MUST explicitly state *which* agent(s) (1, 2, or 3) caused the failure AND *why*, using reasoning found in the "Original Reasoning" context.
5.  **PASS Case Check:** If the final decision is "Approved", the "Recommendation" MUST explicitly state *why* it is a strong application, using reasoning found in the "Original Reasoning" context (e.g., "low credit risk," "no fraud indicators").

**--- IMPORTANT EXCEPTIONS (Read Carefully) ---**

1.  **[NEW] Status Interpretation:**
    * For Agent 1 & 2: The "Status" field 'LOW_RISK' means 'PASS'. The "Status" field 'HIGH_RISK' means 'FAIL'.
    * A recommendation that summarizes 'LOW_RISK' as 'passed' is **PERFECTLY FAITHFUL (Score 5)**.
    * A recommendation that summarizes 'HIGH_RISK' as 'failed' is **PERFECTLY FAITHFUL (Score 5)**.

2.  **On Agent 3 'PASS' Status:** If the "Agent 3 Status" is 'PASS', a high-level summary (e.g., "Agent 3 passed") is considered **PERFECTLY FAITHFUL (Score 5)**.

**--- END OF EXCEPTIONS ---**


**【Original Reasoning (Context)】**

**Agent 1 (Credit Risk) Reasoning:**
Status: {agent_1_status}
Probability: {agent_1_probability}
Reasoning Text: {agent_1_reasoning}

**Agent 2 (Fraud Risk) Reasoning:**
Status: {agent_2_status}
Probability: {agent_2_probability}
Reasoning Text: {agent_2_reasoning}

**Agent 3 (Compliance) Reasoning:**
Status: {agent_3_status}
Failure Details: {agent_3_failures}

**【Recommendation to Evaluate】**

**Agent 4 Recommendation:**
{agent_4_recommendation}

**【Your Evaluation】**
Please assess the faithfulness for EACH agent's reasoning individually, based on all Criteria (1-5) and Exceptions above.

Respond STRICTLY in the following JSON format:
{{
  "agent_1_faithful": (true/false, "Is the summary of Agent 1 faithful to its reasoning? (Use Status Interpretation rule)"),
  "agent_1_score": (1-5, "Faithfulness score for A1's part. 1=Total Hallucination, 5=Perfectly Faithful."),
  "agent_2_faithful": (true/false, "Is the summary of Agent 2 faithful to its reasoning? (Use Status Interpretation rule)"),
  "agent_2_score": (1-5, "Faithfulness score for A2's part."),
  "agent_3_faithful": (true/false, "Is the summary of Agent 3 faithful to its reasoning? (Handle 'PASS' status as instructed)"),
  "agent_3_score": (1-5, "Faithfulness score for A3's part."),
  "overall_score": (1-5, "Overall score for the entire recommendation."),
  "justification": ("A brief explanation for any 'false' or low scores. Pinpoint the specific error based *only* on the rules.")
}}
"""

def evaluate_faithfulness_with_llm_judge(df: pd.DataFrame) -> pd.DataFrame:
    """
    (LLM Judge) Evaluate whether the 'recommendation' is faithful to the 'reasoning' of A1, A2, and A3.
    (This version adds A1/A2 Status fields)
    """
    print("=" * 70)
    print("Agent 4 Recommendation Faithfulness (Granular LLM Judge)")
    print("=" * 70)

    evaluation_results = []
    total_cases = len(df)

    for index, row in df.iterrows():
        applicant_id = row.get('applicant_data.ID', index)
        print(f"--- Evaluating Case ID: {applicant_id} ({index+1}/{total_cases}) ---")

        # 1. Extract all original reasoning

        # --- MODIFICATION START: Extract A1/A2 Status ---

        # A1
        a1_status = row.get('agent1_result.analysis.risk_level', 'N/A')  # [NEW]
        a1_reason = row.get('agent1_result.analysis.reasoning', 'N/A')
        a1_prob = row.get('agent1_result.analysis.default_probability', 'N/A')
        if pd.isna(a1_reason) or a1_reason == 'N/A':
            a1_reason = row.get('agent1_result.reasoning', 'N/A')
        a1_reason = "N/A" if pd.isna(a1_reason) else a1_reason
        a1_status = "N/A" if pd.isna(a1_status) else a1_status

        # A2
        a2_status = row.get('agent2_result.risk_level', 'N/A')  # [NEW]
        a2_reason = row.get('agent2_result.reasoning', 'N/A')
        a2_prob = row.get('agent2_result.fraud_probability', 'N/A')
        a2_reason = "N/A" if pd.isna(a2_reason) else a2_reason
        a2_status = "N/A" if pd.isna(a2_status) else a2_status

        # A4
        a4_recommendation = row.get('agent4_final_report.recommendation', 'N/A')
        a4_recommendation = "N/A" if pd.isna(a4_recommendation) else a4_recommendation

        # A3
        a3_status = row.get('agent3_result.Overall_compliance_status', 'N/A')
        a3_status = "N/A" if pd.isna(a3_status) else a3_status

        a3_failures = "N/A"
        check_details = row.get('agent3_result.check_details', [])
        if not isinstance(check_details, list):
             check_details = []

        if a3_status == 'FAIL':
            fail_reasons = [
                f"{check.get('check')}: {check.get('justification')}"
                for check in check_details if check.get('result') == 'FAIL'
            ]
            a3_failures = "; ".join(fail_reasons) if fail_reasons else "N/A"
        else:
            pass_reasons = [
                f"{check.get('check')}: {check.get('justification')}"
                for check in check_details if check.get('result') == 'PASS'
            ]
            a3_failures = "; ".join(pass_reasons) if pass_reasons else "All checks passed."

        # --- MODIFICATION END ---

        # 2. Build Judge Prompt (insert new status variables)
        prompt = LLM_JUDGE_PROMPT_TEMPLATE.format(
            agent_1_status=a1_status,           # [NEW]
            agent_1_probability=a1_prob,
            agent_1_reasoning=a1_reason,
            agent_2_status=a2_status,           # [NEW]
            agent_2_probability=a2_prob,
            agent_2_reasoning=a2_reason,
            agent_3_status=a3_status,
            agent_3_failures=a3_failures,
            agent_4_recommendation=a4_recommendation
        )

        # 3. Call LLM Judge
        response_text = call_llm_judge(prompt)

        # 4. Parse results
        eval_json = extract_json_from_response(response_text)

        eval_json['applicant_id'] = applicant_id
        evaluation_results.append(eval_json)

        # Print individual result
        print(f"  A1 Score: {eval_json.get('agent_1_score', 'N/A')}, A2 Score: {eval_json.get('agent_2_score', 'N/A')}, A3 Score: {eval_json.get('agent_3_score', 'N/A')}")
        print(f"  Justification: {eval_json.get('justification', 'N/A')}")

    print("\n✓ LLM Judge (granular) evaluation complete.")
    return pd.DataFrame(evaluation_results)
```

```python
# [ Cell 5:  Mention Rate Statistics Function ]

def calculate_mention_rates(df: pd.DataFrame) -> dict:
    """
    (Statistical) Check whether the 'recommendation' field mentions A1, A2, or A3.
    """
    print("Calculating Mention Rates...")
    recommendations = df['agent4_final_report.recommendation'].fillna("").tolist()
    total_cases = len(recommendations)
    if total_cases == 0:
        return {'A1_Rate': 0, 'A2_Rate': 0, 'A3_Rate': 0}

    a1_mentions = 0
    a2_mentions = 0
    a3_mentions = 0

    for text in recommendations:
        lower_text = text.lower()

        # Check mention of A1 ("Agent 1" or "credit")
        if "agent 1" in lower_text or "credit" in lower_text:
            a1_mentions += 1

        # Check mention of A2 ("Agent 2" or "fraud")
        if "agent 2" in lower_text or "fraud" in lower_text:
            a2_mentions += 1

        # Check mention of A3 ("Agent 3" or "compliance")
        if "agent 3" in lower_text or "compliance" in lower_text:
            a3_mentions += 1

    print("✓ Mention Rates calculated.")
    return {
        'A1_Rate': (a1_mentions / total_cases) * 100,
        'A2_Rate': (a2_mentions / total_cases) * 100,
        'A3_Rate': (a3_mentions / total_cases) * 100
    }
```

```python
# [ Cell 6: Run Evaluation and Generate Your Table ]

# --- Run Evaluation ---

# !! IMPORTANT !!
# Make sure this path points to your merge.json file.
file_path = "/content/drive/MyDrive/Gen AI group project/reports/merge.json"  # <-- Ensure the path is correct

# 1. Load data
df = load_data(file_path)

if not df.empty:
    # 2. Run LLM Judge evaluation
    judge_results_df = evaluate_faithfulness_with_llm_judge(df)

    # 3. Run Mention Rate statistics
    mention_rates = calculate_mention_rates(df)

    # --- Display LLM Judge Summary Report (the table you requested) ---
    print("\n" + "=" * 70)
    print("Agent 4 Faithfulness - Summary Table")
    print("=" * 70)

    # 4. Clean and compute scores
    judge_results_df['agent_1_score'] = pd.to_numeric(judge_results_df['agent_1_score'], errors='coerce')
    judge_results_df['agent_2_score'] = pd.to_numeric(judge_results_df['agent_2_score'], errors='coerce')
    judge_results_df['agent_3_score'] = pd.to_numeric(judge_results_df['agent_3_score'], errors='coerce')

    # 4a. Compute Avg_Score (average score for all cases)
    avg_a1_score = judge_results_df['agent_1_score'].mean()
    avg_a2_score = judge_results_df['agent_2_score'].mean()
    avg_a3_score = judge_results_df['agent_3_score'].mean()

    # 4b. Compute Unfaithful_Cases (number of cases marked as False by Judge)
    uf_a1_count = (judge_results_df['agent_1_faithful'] == False).sum()
    uf_a2_count = (judge_results_df['agent_2_faithful'] == False).sum()
    uf_a3_count = (judge_results_df['agent_3_faithful'] == False).sum()

    # 4c. Compute Avg_Unfaithful_Score (average score only among 'false' cases)
    avg_uf_a1_score = judge_results_df[judge_results_df['agent_1_faithful'] == False]['agent_1_score'].mean()
    avg_uf_a2_score = judge_results_df[judge_results_df['agent_2_faithful'] == False]['agent_2_score'].mean()
    avg_uf_a3_score = judge_results_df[judge_results_df['agent_3_faithful'] == False]['agent_3_score'].mean()

    # 5. Build table data
    table_data = [
        {
            'Agent': 'Agent 1',
            'Mention_Rate': f"{mention_rates['A1_Rate']:.1f}%",
            'Avg_Score': f"{avg_a1_score:.2f}",
            'Avg_Unfaithful_Score': f"{avg_uf_a1_score:.2f}",
            'Unfaithful_Cases': int(uf_a1_count)
        },
        {
            'Agent': 'Agent 2',
            'Mention_Rate': f"{mention_rates['A2_Rate']:.1f}%",
            'Avg_Score': f"{avg_a2_score:.2f}",
            'Avg_Unfaithful_Score': f"{avg_uf_a2_score:.2f}",
            'Unfaithful_Cases': int(uf_a2_count)
        },
        {
            'Agent': 'Agent 3',
            'Mention_Rate': f"{mention_rates['A3_Rate']:.1f}%",
            'Avg_Score': f"{avg_a3_score:.2f}",
            'Avg_Unfaithful_Score': f"{avg_uf_a3_score:.2f}",
            'Unfaithful_Cases': int(uf_a3_count)
        }
    ]

    summary_table_df = pd.DataFrame(table_data).set_index('Agent')
    # Replace 'nan' with 'N/A' (if there are no unfaithful cases)
    summary_table_df['Avg_Unfaithful_Score'] = summary_table_df['Avg_Unfaithful_Score'].replace("nan", "N/A")

    # 6. Print table (in Markdown format to mimic the figure)
    print(summary_table_df.to_markdown(numalign="center", stralign="center"))

    # 7. Print text summary (as shown in the figure)
    try:
        # Ensure valid scores (not NaN)
        scores_dict = {
            'Agent 1': avg_a1_score if pd.notna(avg_a1_score) else 0,
            'Agent 2': avg_a2_score if pd.notna(avg_a2_score) else 0,
            'Agent 3': avg_a3_score if pd.notna(avg_a3_score) else 0
        }
        uf_dict = {
            'Agent 1': uf_a1_count,
            'Agent 2': uf_a2_count,
            'Agent 3': uf_a3_count
        }

        # Identify best and worst agents
        best_agent = max(scores_dict, key=scores_dict.get)
        worst_uf_agent = max(uf_dict, key=lambda k: (uf_dict[k], -scores_dict[k]))  # Most unfaithful cases; if tie, choose the one with lower score

        print("\n\n**This table tells you:**\n")
        print(f"✅ **{best_agent} performs the best** - Average {scores_dict[best_agent]:.2f} points, with only {uf_dict[best_agent]} unfaithful cases.")
        print(f"⚠️ **{worst_uf_agent} needs improvement** - {uf_dict[worst_uf_agent]} unfaithful cases (accounting for {uf_dict[worst_uf_agent]/len(df):.1%}).")
        print(f"ℹ️ **Avg_Unfaithful_Score** - When the Judge detects an error, this represents the average severity of that error (lower score = more severe).")

    except Exception as e:
        print(f"Error generating summary text: {e}")

    # 8. Display specific cases flagged as unfaithful by the Judge
    unfaithful_cases = judge_results_df[
        (judge_results_df['agent_1_faithful'] == False) |
        (judge_results_df['agent_2_faithful'] == False) |
        (judge_results_df['agent_3_faithful'] == False)
    ]

    if not unfaithful_cases.empty:
        print("\n--- Cases Flagged as UNFAITHFUL ---")
        # Display detailed columns
        display(unfaithful_cases[['applicant_id', 'agent_1_faithful', 'agent_1_score', 'agent_2_faithful', 'agent_2_score', 'agent_3_faithful', 'agent_3_score', 'justification']])
    else:
        print("\n\n✓ Congratulations! All cases from A1, A2, and A3 were judged as faithful.")
```
